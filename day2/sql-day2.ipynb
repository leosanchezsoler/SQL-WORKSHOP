```python
# Import the necessary libraries

import sqlite3
import pandas as pd
```

```python
# Load the data
data = 'data/chinook.db'

# Use connect()
conn = sqlite3.connect(data)
```

```python
# Create a cursor
cursor = conn.cursor()
```

```python
# Create the query
query = '''
SELECT *
FROM albums
'''
```

```python
# Execute the query
first_query = cursor.execute(query)
```

```python
# Use .fetchall() to display the data
first_query.fetchall()
```

```python
### Now use pandas .read_sql_query
    # Needs 2 arguments:
        # The query
        # The connection
pd.read_sql_query(query, conn)
```

```python
# Make a query with labels from 2 tables
query = '''
SELECT *
FROM customers, invoices
WHERE customers.CustomerID = invoices.CustomerId
'''
pd.read_sql_query(query, conn)
```

```python
# Make a query with labels from 2 tables
query= """
SELECT *
FROM customers
JOIN invoices ON customers.customerid = invoices.customerid
"""
pd.read_sql_query(query, con=conn)
```

```python
query = '''
SELECT c.FirstName, i.InvoiceId, i.InvoiceDate, i.BillingCountry
FROM customers c
JOIN invoices i ON c.CustomerId = i.CustomerId
WHERE c.country = 'Brazil'
'''
pd.read_sql_query(query, con=conn)
```

```python
# Show the name of the customer, the country, the name of the agent and the total
query = '''
SELECT c.FirstName "Customer", c.Country, e.FirstName 'Employee', i.Total 
FROM invoices i
JOIN customers c ON i.CustomerID = c.customerID
JOIN employees e ON e.EmployeeID = c.supportRepID
'''
pd.read_sql_query(query, conn)
```

```python
query = '''
SELECT e.LastName ||' ' || e.FirstName as 'Employee Name', i. * 
FROM employees e
JOIN customers c ON e.EmployeeId = c.SupportRepId
JOIN invoices i ON c.CustomerId = i.CustomerID
'''
pd.read_sql_query(query, conn)
```

```python
# Show every invoice item 
query = '''
SELECT *
FROM tracks t
JOIN invoice_items ii ON ii.trackid = t.trackid
JOIN invoices i ON i.invoiceid = ii.invoiceid
'''
pd.read_sql_query(query, conn)
```

```python
query = '''
SELECT t.name, ii.invoicelineid, i.invoiceid
FROM tracks t
JOIN invoice_items ii ON ii.trackid = t.trackid
JOIN invoices i ON i.invoiceID = ii.invoiceID
'''
pd.read_sql_query(query, conn)
```

```python
query = '''
SELECT t.name 'Song Name', mt.name 'Format', al.Title 'Album Name', g.Name 'Genre'
FROM tracks t
JOIN media_types mt ON t.mediatypeID = mt.mediatypeID
JOIN albums al ON t.albumID = al.albumId
JOIN genres g ON g.genreID = t.genreid
'''
pd.read_sql_query(query, conn)
```

```python
query = '''
SELECT t.Name, t.MediaTypeId, t.AlbumID, t.GenreId
FROM tracks as t
'''
pd.read_sql_query(query, conn)
```

```python
query = '''
SELECT pl.Name, COUNT (*) 'Total'
FROM Playlists pl
JOIN playlist_track pt ON pl.playlistId = pt.playlistID
GROUP BY pt.playlistID
'''
pd.read_sql_query(query, conn)
```

```python
query = '''
SELECT e.FirstName, SUM(i.Total) 'Sales'
FROM employees e
JOIN customers c ON e.EmployeeID = c.SupportRepId
JOIN invoices i ON i.CustomerID = c.CustomerID
GROUP BY e.EmployeeID
'''
pd.read_sql_query(query, conn)
```

```python
query = '''
SELECT e.FirstName, SUM(i.Total) 'Sales'
FROM employees e
JOIN customers c ON e.EmployeeID = c.SupportRepId
JOIN invoices i ON i.CustomerID = c.CustomerID
WHERE i.InvoiceDate LIKE '2009%'
GROUP BY e.EmployeeID
ORDER BY Sales DESC
LIMIT 1 
'''
# With LIMIT, we can select the number of results that will be displayed
pd.read_sql_query(query, conn)
```

```python
query = '''
SELECT ar.Name, SUM(ii.UnitPrice) 'Sales'
FROM artists ar
JOIN albums al ON al.artistID = ar.artistID
JOIN tracks t ON  al.albumID = t.albumID
JOIN invoice_items ii ON t.trackID = ii.TrackID
GROUP BY ar.ArtistID
ORDER BY Sales DESC
LIMIT 3
'''
pd.read_sql_query(query, conn)
```
